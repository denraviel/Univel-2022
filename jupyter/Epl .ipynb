```python
import bs4
import requests
```

```python
soup = bs4.BeautifulSoup
```

```python
response = requests.get("https://www.premierleague.com/clubs")
```

```python

first_soup = soup(response.text).find_all("div", {"class" :"nameContainer"})
first_soup
```

```python
second_soup = soup(response.text).find_all("li")
second_soup
```

```python
for club_name in first_soup:
    
   
    
   
    print(club_name.text.strip(), end = "\n\n")
    

    
    
```

```python
for x in second_soup:
    
    
    link_tag = x.find("a")
    link = link_tag.attrs["href"]
    if  ("http") not in link:
        break
    

    
    print(link)
    
```
